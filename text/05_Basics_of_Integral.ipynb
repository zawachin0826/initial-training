# 積分の基礎

## 不定積分
不定積分は、微分積分学の基本的な概念の一つで，ある関数の原始関数を求める操作を指します。関数 $F(x)$ が関数 $f(x)$ の原始関数であるとは，その関数 $F(x)$ の導関数 $F'(x)$ が，関数 $f(x)$ に一致することを指します。
すなわち，不定積分は，微分の逆の演算であると言えます．

関数 $f(x)$ の $x$ に関する不定積分を、積分記号 $\int$（インテグラル）を用いて，
$$
\int f(x) dx
$$
と表します．

関数 $f(x) = x^2$ の不定積分を求めてみましょう．
これを解くためには，微分したら $x^2$ になる関数を探す必要があります．
第4章で扱った公式
$$
(x^n)' = nx^{n-1}
$$
を思い出すと，多項式を微分すると次数が1下がることがわかります．
よって，微分したら $x^2$ になる関数は，微分のときとは逆に次数を1足して $x^3$ になりそうです．
本当にそうでしょうか？ $x^3$ を微分して確かめてみると，
$$
(x^3)' = 3x^2
$$
となり，$x^2$の3倍になってしまいました．
